```python
#Import libraries
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
#Read data
```

```python
df = pd.read_csv("heart_attack_prediction_dataset.csv")
```

```python
#Cleaning data
```

```python
df = df.drop(columns=["Patient ID"])
df
```

```python
systolic = []
diastolic = []
for i in df["Blood Pressure"]:
  split_data = i.split("/")
  systolic.append(split_data[0])
  diastolic.append(split_data[1])
```

```python
bp_loc = df.columns.get_loc("Blood Pressure")
```

```python
df.insert(bp_loc,"Systolic", systolic)
df.insert(bp_loc,"Diastolic", diastolic)
df
```

```python
df = df.drop(columns=["Blood Pressure"])
```

```python
#Converting strings to 0s and 1s
#Male -> 0
#Female -> 1
#unhealthy -> 0
#healthy -> 1
#average = 2
#Also removing other columns
df = df.drop(columns=["Hemisphere", "Continent", "Country"])
df = df.replace(to_replace=["Male"], value=0)
df = df.replace(to_replace=["Female"], value=1)
df = df.replace(to_replace=["Unhealthy"], value=0)
df = df.replace(to_replace=["Healthy"], value=1)
df = df.replace(to_replace=["Average"], value=2)
```

```python
df.to_csv("cleaned_data.csv")
```

```python
#Splitting into input and output dataset
```

```python
X = df.drop(columns=["Heart Attack Risk"])
y = df["Heart Attack Risk"]
```

```python
#Splitting into test and train datasets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
#Training model and initalizing it
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
#Making predictions
pred = model.predict(X_test)
```

```python
#Calculating accuracy
score = accuracy_score(y_test, pred)
score
```
